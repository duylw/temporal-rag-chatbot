```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma
from dotenv import load_dotenv
load_dotenv()
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-2-preview")
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-preview")
```

```python
chroma_client = Chroma(
    embedding_function=embeddings,
    host="localhost",
    port="8008",
)
```

```python
chroma_client.get(include=["metadatas"])
```

```python
import pickle
with open('../data/vector_data_export.pkl', 'rb') as f:
    loaded_data = pickle.load(f)

# Add to the other_vector_db's underlying collection
chroma_client._collection.add(
    ids=loaded_data['ids'],
    embeddings=loaded_data['embeddings'],
    metadatas=loaded_data['metadatas'],
    documents=loaded_data['documents']
)
```
